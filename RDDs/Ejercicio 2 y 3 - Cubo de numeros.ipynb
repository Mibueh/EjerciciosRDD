```python
class Cubo:
    
    def grafica(self):
        lista = sc.parallelize([5,2,3,4,2,1,5,2,3,4,1,2,5,3]).map(lambda x:(x*x*x)).collect()
        plt.plot(lista)
        plt.show()
        
    def pares(self):
        lista = sc.parallelize([5,2,3,4,2,1,5,2,3,4,1,2,5,3]).map(lambda x:(x*x*x)).collect()
        for x in lista:
            if(x % 2==0):
                print(x,"Es un numero par")
```

```python
prueba=Cubo()
prueba.grafica()
```
